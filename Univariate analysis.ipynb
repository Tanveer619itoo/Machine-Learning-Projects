# Univariate analysis for numerical data

```python
#Histplot  --> Distribution of data
```

```python
sns.histplot(x='Salary',data=df,kde=True)
plt.show()
```

```python
#Boxplot --> find outliers in data
```

```python
sns.boxplot(x='Salary',data=df)
plt.show()
```

```python
#Density plot --> is used to visualize the distribution of a continuous variable.
```

```python
sns.kdeplot(x='Salary',data=df)
plt.show()
```

# Univariate analysis on categorical data

```python
#Frequency count --> count the frequency of categories.
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df['Sex'].value_counts()
```

```python
#Count plot--> count number of categories in a give column
sns.countplot(x='Sex',data=df)
plt.show()
```

```python
#Pie Plot --> used to plot pie chart.
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```
